```python
#DESCRIPTIVE QUESTION
# What is the age distrIbution in the dataset ?

plt.figure(figsize=(10, 4))
sns.histplot(data['Age'], bins=5, color='blue', alpha=1)
plt.title('Age Distribution')
plt.xlabel('Age') 
plt.ylabel('Frequency') 
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data['EstimatedSalary'], bins=5, color='Red', alpha=1)
plt.title('Estimated Salary Distribution')
plt.xlabel('Estimated Salary') 
plt.ylabel('Frequency') 
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data['Purchased'], bins=5, color='Green', alpha=1)
plt.title('Purchased Distribution')
plt.xlabel('Purchased Distribution') 
plt.ylabel('Frequency') 
plt.show()

```

# EXPLORATORY QUESTIONS

```python
sns.boxplot(x='Purchased' , y='Age', data=data)
plt.title('Age vs Purchase')
plt.show()
```

```python
# Age 17 to 52 did not make a purchase and outkiers are present in the set
#age  41 to 60 years made a purchase
```

```python

sns.boxplot(x='Purchased' , y='EstimatedSalary', data=data)
plt.title('Estimated salary vs Purchased')
plt.show()
```

```python
#persons earning 15000 to 77,000 did not make a purchase
#persons earning 41000 to 150,000 made a purchase
```

```python
sns.kdeplot(data[data['Purchased'] == 1]['EstimatedSalary'], label='Purchased')
sns.kdeplot(data[data['Purchased'] == 0]['EstimatedSalary'], label='Not Purchased')
plt.title('Estimated Salary Distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Density') 
plt.legend()
plt.show()                        
```

```python
#persons that purchased more earn more salary than those who didnt make a purchase  
```

```python
#ARE THEY ANY OUTLIERS IN THE DATA SET
sns.boxplot(data=data[['Age', 'EstimatedSalary']]) 
plt.title('Boxplot of Age and Estimated Salary') 
plt.show()
```

```python
#there are no outliers in the estimated salary and age column this simply means that no value in the column is missing
```

```python
#Correlation between AGE AND ESTIMATED SALARY

sns.scatterplot(x='Age', y='EstimatedSalary', data=data)
plt.title('Age vs Estimated Salary')
plt.show()
```

```python
#3D scatter plot of Age, EstimatedSalary and Purchased
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(data['Age'],data['EstimatedSalary'],data['Purchased'],c='r',marker='o')
ax.set_xlabel('Age')
ax.set_ylabel('EstimatedSalary')
ax.set_zlabel('Purchased')
plt.title('3D Scatter Plot of Age, EstimatedSalary and Purchased')
plt.show()
```

```python
#Scatter plot to visualize the distribution
plt.figure(figsize=(10,5))
plt.subplot(131)
plt.scatter(data['Age'],data['EstimatedSalary'],color='blue',alpha=0.8)
plt.title('Age vs EstimatedSalary')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.subplot(132)
plt.scatter(data['Age'],data['Purchased'],color='red',alpha=0.8)
plt.title('Age vs Purchased')
plt.xlabel('Age')
plt.ylabel('Purchased')
plt.subplot(133)
plt.scatter(data['EstimatedSalary'],data['Purchased'],color='brown',alpha=0.8)
plt.title('EstimatedSalary vs Purchased')
plt.xlabel('EstimatedSalary')
plt.ylabel('Purchased')
```
